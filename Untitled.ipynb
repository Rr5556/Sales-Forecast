```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('notebook')
sns.set_palette('Blues_r')
```

# Dataset

```python
df = pd.read_csv('train.csv', encoding = 'unicode escape')
```

```python
df.head()
```

# Data Cleaning

```python
#check any null value
df.isna().sum()
```

```python
df[df['Postal Code'].isna()]
```

```python
df['postal Code']=df['Postal Code'].fillna(5401)
```

```python
df.isna().sum()
```

```python
df.drop(['Row ID'], axis =1, inplace=True)
```

```python
df= df.drop_duplicates()
```

```python
df.info()
```

```python
#converting order_date and Ship_date to datetime data type
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df.info()
```

# Exploratory Data Analysis

```python
# Top region making sales
```

```python
region_sales = df.groupby('Region')['Sales'].sum().round().reset_index().sort_values('Sales', ascending=False)
sns.catplot(data=region_sales, x='Region', y='Sales', kind='bar')
plt.title('Region Sales')
plt.show()
```

```python
#Top 5 cities making Sales
```

```python
city_sales = df.groupby('City')['Sales'].sum().round().reset_index().sort_values('Sales', ascending=False).head()
sns.catplot(data=city_sales, x='City', y='Sales', kind='bar')
plt.title('Top 5 Cities Sales')
plt.xlabel("")
plt.xticks(rotation=90)
plt.show()
```

```python
pvt = df.groupby(['Region', 'Category', 'Sub-Category'])['Sales'].agg(sum)
pvt = pd.DataFrame(pvt)
pvt
```

```python
category = df.groupby('Category')['Sales'].sum().round().reset_index().sort_values(by='Sales', ascending=False).head()
sns.catplot(data=category, x='Category', y='Sales', kind='bar')
plt.title('Categories sales')
plt.xlabel("")
plt.show()
```

```python
#Trend in data set
window_size = 50
rolling_mean = df['Sales'].rolling\
            (window_size).mean()
rolling_mean.plot()
```
